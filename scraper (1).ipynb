```python
file = open("mogtama3.txt", "r")
content = file.read()
```

```python
def extract_fatwas(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    fatwas = []
    
    # Locate all fatwa boxes
    fatwa_boxes = soup.find_all('div', class_='fatwabox')
    
    for box in fatwa_boxes:
        question = ''
        answer = ''
        
        # Extract the question
        question_header = box.find('a', title=True)
        if question_header:
            answer = question_header.get_text(strip=True)
        
        # Extract the answer
        article = box.find('article')
        if article:
            answer_paragraphs = article.find_all('p')
            if answer_paragraphs:
                question = ' '.join(p.get_text(strip=True) for p in answer_paragraphs)
        
        if question and answer:
            fatwas.append({'question': question, 'answer': answer})
    
    return fatwas

html_content = content
fatwas = extract_fatwas(html_content)
for fatwa in fatwas:
    print(f"Question: {fatwa['question']}")
    print(f"Answer: {fatwa['answer']}")
    print("-" * 20)
```
